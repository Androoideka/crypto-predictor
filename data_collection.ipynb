```python
valute_name = "MANA-USD"
file_name = valute_name+".csv"
```

```python
import yfinance as yf

bth = yf.Ticker(valute_name)

hist = bth.history(period="3y",interval="1d")
#print(type(hist))
#print(hist.info())
gg = hist['Close']

#print(gg)
#print(type(gg))
```

```python
save_data(gg,['Date','Close'],file_name)
```

```python
def save_data(series,collums,file_name):
  ress = []
  for index, row in series.items():
    td = {'Date':index,'Close':row}
    ress.append(td)
  
  import csv
  file = 'data/'+file_name
  csv_columns = ['PassengerId','Survived']
  with open(file, 'w') as csvfile:
          writer = csv.DictWriter(csvfile, fieldnames=collums)
          writer.writeheader()
          for data in ress:
              writer.writerow(data)
```

```python
import os
import json
crip = []
for file_name in os.listdir('data/'):
  if file_name.endswith("-USD.csv"):
    crip.append(file_name.replace("-USD.csv",""))

print(len(crip))
with open("cryptocurrencies.txt",'w') as f:
  f.write(json.dumps(crip))
```
