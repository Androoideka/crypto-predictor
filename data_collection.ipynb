```python
import yfinance as yf
import csv
import json
```

```python
def fetch_close(currency_name):
    bth = yf.Ticker(currency_name)
    hist = bth.history(period="2y",interval="1d")
    return hist['Close']
```

```python
def save_data(series,currency_name):
  ress = []
  for index, row in series.items():
    td = {'Date':index,'Close':row}
    ress.append(td)
  
  
  file = 'data/' + currency_name + '.csv'
  csv_columns = ['Date','Close']
  with open(file, 'w') as csvfile:
          writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
          writer.writeheader()
          for data in ress:
              writer.writerow(data)
```

```python
with open("cryptocurrencies.txt",'r') as f:
    currencies = json.load(f)
    for currency in currencies:
        close = fetch_close(currency)
        save_data(close, currency)
```
